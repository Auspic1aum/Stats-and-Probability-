```python
p1, p2, p3, p4, p5, p6, p7, l1, l2, l3, l4 = symbols('p1 p2 p3 p4 p5 p6 p7, l1, l2, l3, l4', real=True)
E_M = -(p1*log(p1) + p2*log(p2) + p3*log(p3) + p4*log(p4) + p5*log(p5) + p6*log(p6) + p7*log(p7))

f1 = p1 + p2 + p3 + p4 + p5 + p6 + p7 - 140 
f2 = p1 + p4 + p5 + p7 - 100 
f3 = p3 + p6 + p7 + p4 -50 
f4 = p2 + p5 + p6 + p7 - 30 


l = E_M + l1*f1 + l2*f2 + l3*f3 + l4*f4 

eq = [diff(l, p1), diff(l, p2), diff(l, p3), diff(l, p4), diff(l, p5), diff(l, p6), diff(l, p7), diff(l, l1), diff(l, l2), diff(l, l3), diff(l, l4)]

solve(eq, (p1, p2, p3, p4, p5, p6, p7, l1, l2, l3, l4))
```

```python
from sympy import symbols, log, diff, solve, Eq

# Define the variables with positive assumptions
p1, p2, p3, p4, p5, p6, p7 = symbols('p1 p2 p3 p4 p5 p6 p7', positive=True)
l1, l2, l3, l4 = symbols('l1 l2 l3 l4')

# Define the entropy function
E_M = -(p1*log(p1) + p2*log(p2) + p3*log(p3) + p4*log(p4) + p5*log(p5) + p6*log(p6) + p7*log(p7))

# Define the constraints
f1 = p1 + p2 + p3 + p4 + p5 + p6 + p7 - 140 
f2 = p1 + p4 + p5 + p7 - 100 
f3 = p3 + p6 + p7 + p4 - 50 
f4 = p2 + p5 + p6 + p7 - 30 

# Define the Lagrangian
L = E_M + l1*f1 + l2*f2 + l3*f3 + l4*f4 

# Compute the partial derivatives (Stationarity conditions)
partial_derivatives = [
    diff(L, p1),
    diff(L, p2),
    diff(L, p3),
    diff(L, p4),
    diff(L, p5),
    diff(L, p6),
    diff(L, p7),
    diff(L, l1),
    diff(L, l2),
    diff(L, l3),
    diff(L, l4)
]

# Create the equations from partial derivatives
equations = [Eq(derivative, 0) for derivative in partial_derivatives]

# Solve the system of equations
solution = solve(equations, (p1, p2, p3, p4, p5, p6, p7, l1, l2, l3, l4), dict=True)

# Print the solution
print(solution)
```
